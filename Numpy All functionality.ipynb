```python
np.random.seed(3)
arr  = np.random.randint(1,10,(3,5))
print(arr)
print('--' * 10)
print(np.sort(arr, axis = 0))


print('--' * 10)
# How we can sort the ray wholy not rowwise or columnwise 

# arr = arr.flatten()
# print(arr)
# arr = np.sort(arr)
# print(arr)
# arr = arr.reshape(3,5)
# print('--' * 10)
# print(arr)


# performing the same above action in the one line this is the beauty of the numpy module 
arr = np.reshape(np.sort(arr.flatten()), arr.shape)
print(arr)
```

# Transposing an array

```python
arr = np.random.randint(1,20,(3,4))
print(arr)
print('--' * 10)
print(arr.T)
```

```python
arr = np.random.randint(1,20,10)
print(arr)
print(np.where(arr % 2 ==0 , 100, arr))
```

```python
arr1 = np.random.randint(1,10,(2,4))
arr2 = np.random.randint(1,10,(2,4))
print(arr1)
print('--' * 10)
print(arr2)
```

```python
print(arr1 + arr2)
```

```python
print(np.concatenate((arr1,arr2), axis = 0))
print('--'* 10)
print(np.concatenate((arr1, arr2), axis = 1))
```

```python
print(arr1)
print('-'*20)
print(arr2)
print('-'*20)
print(np.hstack((arr1,arr2)))
```

```python
print(np.vstack((arr1,arr2)))
```

```python
# Spliting the array

arr = np.random.randint(1,10,(4,4))
print(arr)
print('-' * 20)
print(np.hsplit(arr,2))
print('-'*20)
print(np.vsplit(arr,2))
```
